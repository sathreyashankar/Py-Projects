## Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

## Reading Image Files

```python
image1=cv2.imread('bird1.jpg')
image2=cv2.imread('bird1.jpg')
```

## Creating a Figure with Subplots

```python
ax=plt.subplots(figsize=(15,10))
```

## Image Manipulation Using Bitwise AND, OR and XOR

```python
bitwiseAnd = cv2.bitwise_and(image1,image2)
bitwiseOr = cv2.bitwise_or(image1,image2)
bitwiseXor = cv2.bitwise_xor(image1, image2)
```

## Inverting Image Pixels Using Bitwise NOT


```python
bitwiseNot_img1 = cv2.bitwise_not(image1)
bitwiseNot_img2 = cv2.bitwise_not(image2)
```

##  Visualizing Bitwise AND Operation

```python
plt.subplot(151)
plt.imshow(bitwiseAnd)
```

## Visualizing Bitwise OR Operation

```python
plt.subplot(152)
plt.imshow(bitwiseOr)
```

## Visualizing Bitwise XOR Operation

```python
plt.subplot(153)
plt.imshow(bitwiseXor)
```

## Visualizing Bitwise NOT Operation

```python
plt.subplot(154)
plt.imshow(bitwiseNot_img1)
plt.subplot(155)
plt.imshow(bitwiseNot_img2)
```
